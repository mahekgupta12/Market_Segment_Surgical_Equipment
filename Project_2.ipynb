```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = '/content/Health_Plan_Prior_Authorization_Data.csv'
df = pd.read_csv(file_path)
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
plt.figure(figsize=(12, 6))
```

```python
plt.subplot(2, 2, 1)
sns.histplot(df['Carrier'], kde=True)
plt.title('Distribution of Carrier')
```

```python
plt.subplot(2, 2, 2)
sns.countplot(data=df, x='Year')
plt.title('Count of Year')
```

```python
plt.subplot(2, 2, 3)
sns.boxplot(data=df, x='Year', y='Carrier')
plt.title('Box plot of Carrier by Year')
```

```python
plt.subplot(2, 2, 4)
numeric_df = df.select_dtypes(include=['float64', 'int64'])
corr = numeric_df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
```

```python
plt.tight_layout()
plt.show()
```
